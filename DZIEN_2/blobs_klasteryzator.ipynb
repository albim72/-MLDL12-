```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import kmeans_plusplus
```

```python
n_samples = 15_000
n_components = 8
```

```python
X,y_true = make_blobs(n_samples=n_samples,centers=n_components,cluster_std = 0.7,random_state=1)
X = X[:,::-1]
```

```python
centers_init, indices = kmeans_plusplus(X,n_clusters=8,random_state=1)
```

```python
plt.figure()
colors = ["red","orange","m","#AA55B7","#11FF99","c","#986432","g"]
for k,col in enumerate(colors):
    cluster_data = y_true == k
    plt.scatter(X[cluster_data,0],X[cluster_data,1],c=col,marker=".",s=10)

plt.scatter(centers_init[:,0],centers_init[:,1],c="b",s=60)
plt.title("algorytm k-means")
plt.xticks([])
plt.yticks([])
plt.show()
```
